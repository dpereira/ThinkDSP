```python
import thinkdsp, numpy

ss = thinkdsp.SinSignal(freq=4400)

cs = ss + thinkdsp.SinSignal(freq=4800) + thinkdsp.SinSignal(freq=4900) + thinkdsp.SinSignal(freq=5050)  #+ thinkdsp.SinSignal(freq=3001)

w = cs.evaluate(numpy.linspace(0, 1, 1000000))
wave = thinkdsp.Wave(w)
spectrum = wave.make_spectrum()
spectrum.plot()

wave.make_audio()
```

```python
import thinkplot
w = thinkdsp.read_wave("scissors.wav")
s = w.make_spectrum()
s.plot(high=500)
thinkplot.show()

w1 = thinkdsp.read_wave("piano_a4.wav")
s = w1.make_spectrum()
s.plot(high=1000)
thinkplot.show()

w2 = thinkdsp.read_wave("piano_a5.wav")
s = w2.make_spectrum()
s.plot(high=2000)
thinkplot.show()

w3 = w1 + w2 
s = w3.make_spectrum()
s.plot(high=2000)
thinkplot.show()

w3.make_audio()
```

```python
import matplotlib.pyplot as plt
import thinkplot

for i in range(10):
    print(i*0.4)
    w = thinkdsp.read_wave("piano.wav").segment(start=i*0.4,duration=0.4)
    s = w.make_spectrum()
    s.plot(high=2000)
    thinkplot.show(title=str(i*0.4))
```
